```python
import numpy as np
import pandas as pd
# read external dataset
df = pd.read_csv('../data/raw/new york 2016-01-01 to 2017-12-31.csv')
df2 = pd.read_csv('../data/raw/new york 2018-01-01 to 2019-12-31.csv')
```

```python
# append 2018-2019 weather data to 2016-2017 weather external data
df = df.append(df2)[["datetime","tempmin","tempmax","temp","precip","precipcover","windgust","windspeed",
                     "dew","humidity","snow","snowdepth","cloudcover","visibility"]]
```

```python
df
```

```python
# read taxi data after preprocessed
cabs = pd.read_csv('../data/curated/cleaned_data_new.csv')
```

```python
# merge two data tables together, merge on data column from each table
final = cabs.merge(df,left_on='date', right_on='datetime').drop('datetime', axis = 1)
final
```

```python
# to check if there are null values contain in the merged dataset
final.isnull().sum()
```

```python
# suspecting the null values in windgust is due to no wind on that day
# hence, replacing null values by 0
final['windgust'].fillna(0, inplace = True)
```

```python
final.describe()
```

```python
# store it to CSV file
final.to_csv('../data/curated/final_data_new')
```
