```python
import pandas as pd
import os
import numpy as np
```

```python
X_train = pd.read_csv('../data/curated/X_train')
y_train = pd.read_csv('../data/curated/y_train')
X_val = pd.read_csv('../data/curated/X_val')
y_val = pd.read_csv('../data/curated/y_val')
X_test = pd.read_csv('../data/curated/X_test')
y_test = pd.read_csv('../data/curated/y_test')
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
X_train = X_train[X_train.columns[2:]]
X_val = X_val[X_val.columns[2:]]
X_test = X_test[X_test.columns[2:]]
```

```python
X_train
```

```python
X_val
```

```python
y_train = y_train[y_train.columns[1:]]
y_val = y_val[y_val.columns[1:]]
y_test = y_test[y_test.columns[1:]]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
rf_param_group = {
    "n_estimators": (50,60),
    "max_depth": range(10, 80, 10),
#    "min_samples_leaf": range(1, 10, 1),
#    "min_samples_split": range(2, 10, 1),
    "max_features": (0.5,0.75),
    "max_samples" :(0.5,0.75),
    "ccp_alpha" :(0,0.01,0.001,1,10)
}

rf_estimator = RandomForestRegressor()
grid_search = GridSearchCV(estimator=rf_estimator, param_grid=rf_param_group, cv=5, n_jobs=-1)
```

```python
import time
t0 = time.time()
grid_search.fit(X_train, y_train)
print("Execute time of grid search: {} s".format(round(time.time()-t0, 3)))
print("Best params: ", grid_search.best_params_)
```

```python
t0 = time.time()
clf_rf = RandomForestRegressor(n_estimators=60, 
                                max_depth=30, max_features=0.75,
                                ccp_alpha = 0.001, max_samples = 0.5).fit(X_train, y_train)
print("Execute time of Random forest: {} s".format(round(time.time()-t0, 3)))
```

```python
from sklearn import metrics
rf_train_predict = clf_rf.predict(X_train)
print('Mean Absolute Error For Training Set:{:.3f}'.format(metrics.mean_absolute_error(y_train, rf_train_predict)))
print('Mean Squared Error For Training Set:{:.3f}'.format(metrics.mean_squared_error(y_train, rf_train_predict)))
print('Root Mean Squared Error For Training Set:{:.3f}'.format(np.sqrt(metrics.mean_squared_error(y_train, rf_train_predict))))
```

```python
rf_val_predict = clf_rf.predict(X_val)
print('Mean Absolute Error For Validation Set:{:.3f}'.format(metrics.mean_absolute_error(y_val, rf_val_predict)))
print('Mean Squared Error For Validation Set:{:.3f}'.format(metrics.mean_squared_error(y_val, rf_val_predict)))
print('Root Mean Squared Error For Validation Set:{:.3f}'.format(np.sqrt(metrics.mean_squared_error(y_val, rf_val_predict))))
```

```python
rf_test_predict = clf_rf.predict(X_test)
print('Mean Absolute Error For Testing Set:{:.3f}'.format(metrics.mean_absolute_error(y_test, rf_test_predict)))
print('Mean Squared Error For Testing Set:{:.3f}'.format(metrics.mean_squared_error(y_test, rf_test_predict)))
print('Root Mean Squared Error For Testing Set:{:.3f}'.format(np.sqrt(metrics.mean_squared_error(y_test, rf_test_predict))))
```

```python
# save the prediction result for testing dataset for mapping section
test_pre = pd.DataFrame(rf_test_predict, columns = ['predicted_income_rate'])
test_pre.to_csv('../data/curated/RFR_test_prediction.csv')
```
