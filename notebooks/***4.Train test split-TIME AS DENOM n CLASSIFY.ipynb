```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read in the merged table of weather and taxi data
df = pd.read_csv('../data/curated/final_data_new')
```

```python
df.describe()
```

```python
# split data into response(income_rate) and the remainders as predictors
X = df[['PULocationID','total_adjusted_amount','trip_distance','fare_amount','tip_amount',
       "tempmin","tempmax","temp","precip","precipcover","windgust","windspeed",
                     "dew","humidity","snow","snowdepth","cloudcover","visibility"]]
```

```python
y = df[['income_rate']]
```

```python
# split training set and validation set chronologically
X_train, X_test, y_train, y_test = train_test_split(X, y,
    test_size=0.4, shuffle = False)
```

```python
# equally split the validation set and testing set, shuffle split to make sure more equal distributions between two sets
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, 
    test_size=0.5, shuffle = True, random_state= 42)
```

```python
# to remove PULocationID
train = X_train[X_train.columns[1:]]
```

```python
# Standardise by sklearn
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X_train.iloc[:, 1:] = SS.fit_transform(X_train.iloc[:, 1:])
```

```python
# store train/val/test set
X_train.to_csv('../data/curated/X_train')
y_train.to_csv('../data/curated/y_train')
X_val.to_csv('../data/curated/X_val')
y_val.to_csv('../data/curated/y_val')
X_test.to_csv('../data/curated/X_test')
y_test.to_csv('../data/curated/y_test')
```
